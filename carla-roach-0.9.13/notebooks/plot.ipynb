```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = [
    {
        "Road Type": "Highway",
        "Number of Collision": 39,
    },
    {
        "Road Type": "Intersection",
        "Number of Collision": 20,
    },
    {
        "Road Type": "Merge Lane",
        "Number of Collision": 10,
    },
    {
        "Road Type": "Exit Lane",
        "Number of Collision": 14,
    },

]


df = pd.DataFrame(data)
print(df['Number of Collision'].sum())
plt.figure(figsize=(10, 5))
sns.barplot(x="Road Type", y="Number of Collision", data=df)
plt.xlabel("Road Type")
plt.ylabel("Number of Accidents")
plt.title("Number of Accidents by Road Type")
plt.show()
```
